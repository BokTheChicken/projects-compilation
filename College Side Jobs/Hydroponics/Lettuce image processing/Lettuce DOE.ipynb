```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

# Image Preprocessing

```python
#Sample Images
w=10
h=10
fig=plt.figure(figsize=(10, 12))
columns = 2
rows = 5
x=0
for i in range(1, columns*rows +1):
    x=x+1
    image = cv2.imread("600x450_0"+str(x)+".jpg")
    image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    fig.add_subplot(rows, columns, i)
    plt.imshow(image_RGB)
plt.show()
```

```python
#Grayscale

w=10
h=10
fig=plt.figure(figsize=(10, 12))
columns = 2
rows = 5
x=0
for i in range(1, columns*rows +1):
    x=x+1
    image = cv2.imread("600x450_0"+str(x)+".jpg")
    image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image_gray = cv2.cvtColor(image_RGB, cv2.COLOR_RGB2GRAY)
    fig.add_subplot(rows, columns, i)
    plt.imshow(image_gray, cmap='gray')
plt.show()
```

```python
#Image Smoothening

w=10
h=10
fig=plt.figure(figsize=(10, 12))
columns = 2
rows = 5
x=0
for i in range(1, columns*rows +1):
    x=x+1
    image = cv2.imread("600x450_0"+str(x)+".jpg")
    image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image_gray = cv2.cvtColor(image_RGB, cv2.COLOR_RGB2GRAY)
    median_blur = cv2.medianBlur(image_gray, 9)
    fig.add_subplot(rows, columns, i)
    plt.imshow(median_blur, cmap='gray')
plt.show()
```

# Laplacian

```python
#Laplacian
import time

w=10
h=10
fig=plt.figure(figsize=(10, 12))
columns = 2
rows = 5
start = time.time()
x=0
for i in range(1, columns*rows +1):
    x=x+1
    image = cv2.imread("600x450_0"+str(x)+".jpg")
    image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image_gray = cv2.cvtColor(image_RGB, cv2.COLOR_RGB2GRAY)
    median_blur = cv2.medianBlur(image_gray, 9)
    laplacian_1 = cv2.Laplacian(median_blur, cv2.CV_8U, -1)
    fig.add_subplot(rows, columns, i)
    plt.imshow(laplacian_1, cmap='gray')
plt.show()
stop = time.time()
total = stop - start
print("Laplacian Time Delay:")
print(str(total)+"secs")
```

# SobelX

```python
#SobelX
w=10
h=10
fig=plt.figure(figsize=(10, 12))
columns = 2
rows = 5
start = time.time()
x=0
for i in range(1, columns*rows +1):
    x=x+1
    image = cv2.imread("600x450_0"+str(x)+".jpg")
    image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image_gray = cv2.cvtColor(image_RGB, cv2.COLOR_RGB2GRAY)
    median_blur = cv2.medianBlur(image_gray, 9)
    sobelx = cv2.Sobel(median_blur,cv2.CV_64F,1,0,ksize=5)
    sobelx_165 = np.absolute(sobelx)
    sobelx_8U = np.uint8(sobelx_165)
    fig.add_subplot(rows, columns, i)
    plt.imshow(sobelx_8U, cmap='gray')
plt.show()

stop = time.time()
total = stop - start
print("SobelX Time Delay:")
print(str(total)+"secs")
```

# Canny

```python
#Canny

start = time.time()
w=10
h=10
fig=plt.figure(figsize=(10, 12))
columns = 2
rows = 5
start = time.time()
x=0
for i in range(1, columns*rows +1):
    x=x+1
    image = cv2.imread("600x450_0"+str(x)+".jpg")
    image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image_gray = cv2.cvtColor(image_RGB, cv2.COLOR_RGB2GRAY)
    median_blur = cv2.medianBlur(image_gray, 9)
    edges = cv2.Canny(image_gray,20,75)
    fig.add_subplot(rows, columns, i)
    plt.imshow(edges, cmap='gray')
plt.show()
stop = time.time()
total = stop - start
print("Canny Time Delay:")
print(str(total)+"secs")
```
